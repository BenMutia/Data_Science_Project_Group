# Group Assignment (Violin)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv("1500000 Sales Records (1).csv")
df.head(5)
```

## Violin Plot (Total Profit vs Sales Channel)

```python
sns.violinplot(x='Sales Channel', y='Total Profit', hue='Sales Channel',legend=False, data=df, palette='coolwarm')

plt.title("spread of Total Profit across the different Sales Channels", fontsize=14)
plt.xlabel("Sales Channels")
plt.ylabel("Total Profit")
plt.show()
```

## Violin Plot (Total Revenue vs Sales Channel)

```python
sns.violinplot(x='Sales Channel', y='Total Revenue', hue='Sales Channel',legend=False, data=df, palette='coolwarm')

plt.title("spread of Total Revenue across the different Sales Channels", fontsize=14)
plt.xlabel("Sales Channels")
plt.ylabel("Total Revenue")
plt.show()
```

```python
df.columns
```

```python
Selected_Features = df[['Units Sold', 'Unit Price', 'Unit Cost', 'Total Revenue', 'Total Cost', 'Total Profit']]
Corr_matrix = Selected_Features.corr()
dfc = pd.DataFrame(Corr_matrix)

fig, ax = plt.subplots(figsize=(10, 6))
ax.axis('tight')
ax.axis('off')
table = ax.table(cellText=dfc.values, colLabels=dfc.columns, rowLabels=dfc.columns,cellLoc='center', loc='center')
plt.show()

plt.figure(figsize=(8, 6))
sns.heatmap(Corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap for Selected Features", fontsize=14)
plt.show()
```

```python
sns.histplot(df["Total Revenue"], bins=30, kde=True)
plt.title("Distribution of Revenue")
plt.show()
```

```python
plt.figure(figsize=(12,4))
sns.boxplot(x = 'Item Type',y = 'Total Revenue', hue = 'Item Type', legend =False, data = df, palette='viridis')
plt.title('Distribution of Total Revenue Accross Item Types')
plt.xlabel ("Item Type")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.show()
```

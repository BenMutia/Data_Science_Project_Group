```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("1500000 Sales Records (1).csv")
df.head()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Units Sold'], bins=20, alpha=0.7, kde=True)

plt.title("Distribution of Units Sold")
plt.xlabel("Units Sold")
plt.ylabel("Frequency")

plt.show()
```
